```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

city_path = 'data/city_data.csv'
ride_path = 'data/ride_data.csv'
```

```python
city_df = pd.read_csv(city_path)
ride_df = pd.read_csv(ride_path)

pyber_df = pd.merge(ride_df, city_df, on='city', how='outer')

sum_df = pyber_df.groupby('city').sum()
count_df = pyber_df.groupby('city').count()
plot_df = city_df

avg_fare = []
avg_fare.append(sum_df['fare'] / count_df['fare'])

avg_fare = pd.DataFrame(avg_fare)
avg_fare = avg_fare.T

plot_df = pd.merge(plot_df, avg_fare, on='city')

ride_count = count_df['date']
ride_count_df = pd.DataFrame(ride_count)
plot_df = pd.merge(plot_df, ride_count_df, on='city')


plot_df = plot_df.rename(columns={'city':'City',
                        'driver_count':'Total Drivers',
                        'type':'Type',
                        'fare':'Average Fare',
                        'date':'Total Rides'
                       })
#plot_df['Average Fare'] = plot_df['Average Fare'].map('${:,.2f}'.format)

conditions = [
    (plot_df['Type'] == 'Urban'),
    (plot_df['Type'] == 'Suburban'),
    (plot_df['Type'] == 'Rural')]
choices = ['Coral', 'SkyBlue', 'Gold']
plot_df['color'] = np.select(conditions, choices)


urban_df = plot_df[plot_df['Type']=='Urban']
suburban_df = plot_df[plot_df['Type']=='Suburban']
rural_df = plot_df[plot_df['Type']=='Rural']
```

```python
x = urban_df['Total Rides']
y = urban_df['Average Fare']
s = urban_df['Total Drivers']
fc = urban_df['color']
plt.scatter(x, y, s=s*3, alpha=0.7, facecolors=fc, edgecolors='black')

x = suburban_df['Total Rides']
y = suburban_df['Average Fare']
s = suburban_df['Total Drivers']
fc = suburban_df['color']
plt.scatter(x, y, s=s*3, alpha=0.7, facecolors=fc, edgecolors='black')

x = rural_df['Total Rides']
y = rural_df['Average Fare']
s = rural_df['Total Drivers']
fc = rural_df['color']
plt.scatter(x, y, s=s*3, alpha=0.7, facecolors=fc, edgecolors='black')

plt.title("Pyber Ride Sharing Data")
plt.xlabel("Total Number of Rides")
plt.ylabel("Average Fare of Rides")
plt.legend(['Urban', 'Suburban', 'Rural'])
plt.grid(True)
plt.text(42, 40, 'Note: Circle Size Correlates with Driver Count per City')
```

```python
fares_df = pyber_df.groupby('type').sum()
colors = ['gold', 'skyblue', 'coral']
plt.pie(fares_df['fare'], labels=fares_df.index, colors=colors, autopct="%1.1f%%")
plt.title('Percent of Total Fares in Each City Type')
```

```python
rides_df = pyber_df.groupby('type').count()
colors = ['gold', 'skyblue', 'coral']
plt.pie(rides_df['city'], labels=fares_df.index, colors=colors, autopct='%1.1f%%')
plt.title('Percent of Total Rides in Each City Type')
```

```python
drivers_df = city_df.groupby('type').sum()
colors = ['gold', 'skyblue', 'coral']
plt.pie(drivers_df['driver_count'], labels=drivers_df.index, colors=colors, autopct='%1.1f%%')
plt.title('Percent of Total Drivers in Each City Type')
```
